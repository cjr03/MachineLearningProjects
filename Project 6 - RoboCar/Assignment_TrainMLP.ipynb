```python
import numpy as np
import numpy.matlib
import pandas as pd

InputNodes = 3
HiddenNodes = 2
OutputNodes = 2

Input = np.array([[0,0,0],
                  [0,0,1],
                  [0,1,0],
                  [1,0,0]])

Target = np.array([[1,1],
                   [0,0],
                   [1,0],
                   [0,1]])

Hidden = np.zeros((HiddenNodes,1))
Output = np.zeros((OutputNodes,1))

# Initialize the weights randomly with a seed
rng = np.random.default_rng(seed=12)
HiddenWeights = rng.random((HiddenNodes, InputNodes+1))
OutputWeights = rng.random((OutputNodes, HiddenNodes+1))

## Your code goes here ...
LearningRate = 2.0
NumPatterns = Input.shape[0]
MaxEpochs = 15000
ErrorThreshold = 0.005

def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

for epoch in range(MaxEpochs):
    # Randomize the dataset each epoch
    indices = rng.permutation(NumPatterns)
    Input_shuffled = Input[indices]
    Target_shuffled = Target[indices]

    Error = 0.0
    for p in range(NumPatterns):
        # Forward pass - Hidden layer
        InputWithBias = np.append(Input_shuffled[p], 1.0)
        HiddenAccum = np.dot(HiddenWeights, InputWithBias)
        HiddenOutput = sigmoid(HiddenAccum)

        # Forward pass - Output layer
        HiddenWithBias = np.append(HiddenOutput, 1.0)
        OutputAccum = np.dot(OutputWeights, HiddenWithBias)
        OutputOutput = sigmoid(OutputAccum)

        # Calculate error
        Error += 0.5 * np.sum((Target_shuffled[p] - OutputOutput) ** 2)

        # Backpropagation - Output layer deltas
        OutputDelta = (Target_shuffled[p] - OutputOutput) * OutputOutput * (1.0 - OutputOutput)

        # Backpropagation - Hidden layer deltas
        HiddenDelta = HiddenOutput * (1.0 - HiddenOutput) * np.dot(OutputWeights[:, :HiddenNodes].T, OutputDelta)

        # Update Output weights
        for j in range(OutputNodes):
            OutputWeights[j, :] += LearningRate * OutputDelta[j] * HiddenWithBias

        # Update Hidden weights
        for j in range(HiddenNodes):
            HiddenWeights[j, :] += LearningRate * HiddenDelta[j] * InputWithBias

    if Error < ErrorThreshold:
        print(f'Converged at epoch {epoch}, Error: {Error}')
        break

if epoch == MaxEpochs - 1:
    print(f'Reached max epochs {MaxEpochs}, Error: {Error}')

# Print Hidden Weights and Output Weights
for i in range(HiddenNodes):
    print('HiddenWeights[{}]: {}'.format(i, HiddenWeights[i,:]))
for i in range(OutputNodes):
    print('OutputWeights[{}]: {}'.format(i, OutputWeights[i,:]))

# Save Hidden Weights and Output Weights in .csv file
df = pd.concat([pd.DataFrame(HiddenWeights), pd.DataFrame(OutputWeights)])
df.to_csv('MLP_weights.csv', index=False, header=False)

# Download to local directory
from google.colab import files
files.download('MLP_weights.csv')
```
